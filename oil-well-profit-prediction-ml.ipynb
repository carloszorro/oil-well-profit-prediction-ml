# Objetivo

Encontrar los mejores lugares donde abrir 200 pozos nuevos de petróleo, mediante las diferentes herramientas aprendidas a lo largo del Sprint, se buscará abordar la situación y demostrarla mediante modelos prácticos.

```python
# importación de librerias
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from numpy.random import RandomState

state = RandomState(9172124)

# Cargar los datos
data_0 = pd.read_csv('/datasets/geo_data_0.csv')
data_1 = pd.read_csv('/datasets/geo_data_1.csv')
data_2 = pd.read_csv('/datasets/geo_data_2.csv')

# Verificar los datos
print(data_0.head())
print(data_1.head())
print(data_2.head())

print(data_0.info())
print(data_1.info())
data_2.info()
```

```python
#Entrenamiento y prueba del modelo

def train_and_evaluate(data):
    features = data[['f0', 'f1', 'f2']]
    target = data['product']
    features_train, features_valid, target_train, target_valid = train_test_split(features, target, test_size=0.25, random_state=state)

    model = LinearRegression()
    model.fit(features_train, target_train)

    predictions = model.predict(features_valid)
    rmse = np.sqrt(mean_squared_error(target_valid, predictions))
    mean_predictions = np.mean(predictions)

    return predictions, target_valid, mean_predictions, rmse, features_valid.index

# Evaluación para cada región
predictions_0, target_valid_0, mean_pred_0, rmse_0, val_indices_0 = train_and_evaluate(data_0)
predictions_1, target_valid_1, mean_pred_1, rmse_1, val_indices_1 = train_and_evaluate(data_1)
predictions_2, target_valid_2, mean_pred_2, rmse_2, val_indices_2 = train_and_evaluate(data_2)

print(f"Región 0: Media de reservas: {mean_pred_0}, RMSE: {rmse_0}")
print(f"Región 1: Media de reservas: {mean_pred_1}, RMSE: {rmse_1}")
print(f"Región 2: Media de reservas: {mean_pred_2}, RMSE: {rmse_2}")
```

## Análisis
La región 2 es la que presenta la mayor media de reservas de barriles (94.46 miles de barriles) seguida de la región 1 (92.46 miles de barriles) y, finalmente, la región 1 (68.29 miles de barriles).
Basandome en los resultados obtenidos se podría identificar que la región 2 es la que mejor potencial de producción podría tener en comparación a las demás regiones.

Ahora bien, al tener presente que si el RMSE es más bajo quiere decir que las predicciones son más cercanas a los valores reales, teniendo en cuenta lo anterior, se puede inferir que en la región 1 es donde más precisa es la predicción, ya sea por la homogeneidad de sus datos o simplemente que el modelo elegido se ajusta mejor a este escenario.

En este punto, a mi modo de ver si se quiere enfocar en las reservas la región 2 es el objetivo, de lo contrario si se desea ser un poco más preciso y cercano al modelo, los datos más confiables vendrían de la región 1.

```python
# Cálculo de ganancias
budget = 100000000
wells = 200
profit_per_barrel = 4.5
profit_per_product = 4500

min_profit_per_well = budget / wells

print(f"Reservas necesarias por pozo: {min_profit_per_well / profit_per_product} unidades")
print(f"Reservas promedio Región 0: {mean_pred_0}, Región 1: {mean_pred_1}, Región 2: {mean_pred_2}")
```

```python
# Cálculo de ganancias para conjunto de 200 pozos

def calculate_profit(data, predictions, val_indices, budget=100_000_000, revenue_per_unit=4500):
    # Añadir predicciones al DataFrame en los índices correctos
    data['predicted_product'] = np.nan  # Inicializar la columna
    data.loc[val_indices, 'predicted_product'] = predictions  # Asignar las predicciones

    # Seleccionar los 200 pozos con las predicciones más altas
    top_wells = data.nlargest(200, 'predicted_product')
    total_reserves = top_wells['predicted_product'].sum()
    profit = total_reserves * revenue_per_unit - budget
    return profit, total_reserves

profit_0, total_reserves_0 = calculate_profit(data_0, predictions_0, val_indices_0)
profit_1, total_reserves_1 = calculate_profit(data_1, predictions_1, val_indices_1)
profit_2, total_reserves_2 = calculate_profit(data_2, predictions_2, val_indices_2)

print(f"Ganancia potencial Región 0: {profit_0}")
print(f"Ganancia potencial Región 1: {profit_1}")
print(f"Ganancia potencial Región 2: {profit_2}")
```

Se puede evidenciar que las mayores ganancias se obtienen en la región 0, no obstante, teniendo en cuenta los resultados anteriores si el ejercicio se enfoca por el tema de reservas la región 2 es hasta el momento la mejor opción.

```python
# Calculo de riesgos y ganancias para cada región

# Función para realizar bootstrapping
def bootstrap_profit(predictions, num_samples=1000, budget=100_000_000, revenue_per_unit=4500):
    profits = []
    for _ in range(num_samples):
        sample = np.random.choice(predictions, size=len(predictions), replace=True)
        total_reserves = np.sum(sample)
        profit = total_reserves * revenue_per_unit - budget
        profits.append(profit)
    return profits

# Función para calcular estadísticas
def calculate_statistics(profits):
    mean_profit = np.mean(profits)
    lower_bound = np.percentile(profits, 2.5)
    upper_bound = np.percentile(profits, 97.5)
    risk_of_loss = np.mean(np.array(profits) < 0) * 100
    return mean_profit, (lower_bound, upper_bound), risk_of_loss

# Realizar bootstrapping para cada región
profits_0 = bootstrap_profit(predictions_0)
profits_1 = bootstrap_profit(predictions_1)
profits_2 = bootstrap_profit(predictions_2)

# Calcular estadísticas para cada región
stats_0 = calculate_statistics(profits_0)
stats_1 = calculate_statistics(profits_1)
stats_2 = calculate_statistics(profits_2)

# Imprimir resultados
print(f"Región 0: Beneficio promedio: {stats_0[0]}, Intervalo de confianza: {stats_0[1]}, Riesgo de pérdidas: {stats_0[2]}%")
print(f"Región 1: Beneficio promedio: {stats_1[0]}, Intervalo de confianza: {stats_1[1]}, Riesgo de pérdidas: {stats_1[2]}%")
print(f"Región 2: Beneficio promedio: {stats_2[0]}, Intervalo de confianza: {stats_2[1]}, Riesgo de pérdidas: {stats_2[2]}%")
```
